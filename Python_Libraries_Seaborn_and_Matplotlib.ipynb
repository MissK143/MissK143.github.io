````python
# Let's create our first plot—a simple line graph representing yearly data.
import matplotlib.pyplot as plt

# Sample data: Yearly average temperatures
years = [2018, 2019, 2020, 2021, 2022]
temperatures = [58.6, 57.9, 59.1, 60.2, 58.7]

plt.figure(figsize=(10, 5))  # Create a figure object with a specified size
plt.plot(years, temperatures, marker='o')  # Plot a line graph
plt.title('Yearly Average Temperatures')  # Add a title
plt.xlabel('Year')  # Add an x-label
plt.ylabel('Temperature (F)')  # Add a y-label
plt.grid(True)  # Add a grid for better readability
plt.show()  # Display the plot
````

## Core Concepts of Matplotlib

Matplotlib is a powerful plotting library that offers extensive tools for visualizing data in Python. One of the fundamental concepts in Matplotlib is the distinction between figures and axes, which are essential to understand before diving into the creation of various types of plots.

### Figures and Axes: The Canvas and Drawing Area

- **Figure:** Think of a figure as the overall window or page where your graphics are drawn. A figure can have multiple axes but at least one.
- **Axes:** These are what we commonly think of as 'plots.' An axes actually represents an individual plot or graph, even though we can have multiple axes (plots) in a single figure.

Here's how you can visualize this relationship: A figure is like a canvas, and each axes is a part of that canvas on which you draw your specific graph.

## Plotting: Creating Various Types of Plots

Matplotlib allows you to create a wide variety of plots. Here, we'll cover the basics: line plots, bar charts, histograms, scatter plots, and pie charts.

### Line Plots

Used primarily to show trends over time.

### Bar Charts

### Histograms

## Customizations: Enhancing Visual Appeal

Matplotlib provides numerous ways to customize your plots to make them more informative and appealing.

- **Titles, Labels, Legends, and Annotations:**
  - **Titles** and **Labels** help explain what the graph is about.
  - **Legends** guide the reader on what each plot element represents.
  - **Annotations** can highlight specific points and add text at these points.


````python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

# Creating a simple line plot
plt.figure(figsize=(8, 4))
plt.errorbar(x, y, yerr=dy, fmt='.k')
plt.title('Simple Line Plot Example')
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.legend(['Data'], loc='best')
plt.grid(True)
plt.show()
````

````python
# Creating a bar chart
categories = ['Category 1', 'Category 2', 'Category 3']
values = [10, 15, 7]
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='blue')
plt.title('Basic Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
````

````python
# Creating a histogram
data = np.random.normal(0, 1, size=1000)
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

````python
# Creating a scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(6, 4))
plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.show()
````

````python
# Creating a pie chart
sizes = [215, 130, 245, 210]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
````

## Advanced Matplotlib

As we advance further into Matplotlib, we'll explore sophisticated techniques that enhance the utility and aesthetics of your plots. These include creating multiple plots within a single figure, applying various styling and themes to customize the appearance, and saving your plots for use in different media. Let's delve into these topics to refine your visualization skills.

### Subplots: Creating Multiple Plots in One Figure

Subplots are an essential feature in Matplotlib, allowing you to organize multiple axes (plots) in a single figure systematically. This is particularly useful for comparing different datasets or aspects of data side-by-side.

#### How Subplots Work:

- **Grid System:** Subplots are arranged in a grid. You specify the number of rows and columns to create the grid layout.
- **Indexing:** Each subplot in the grid is indexed starting from 1 (not 0), moving from left to right and top to bottom.
- **Flexibility:** You can have different sized subplots within the same figure to emphasize particular data.

### Styling and Themes: Using Predefined Styles

Matplotlib offers a variety of predefined styles that can significantly alter the aesthetics of your plots with a single line of code. These styles are inspired by well-known plotting tools and aesthetics from other programming languages.

#### Benefits of Using Styles:

- **Consistency:** Styles ensure that all your plots have a consistent look and feel.
- **Aesthetics:** Quickly improve the visual appeal of your plots to suit presentations or publications.
- **Customization:** While there are several predefined styles, you can also create and use your own custom styles.

### Saving Figures: How to Save Plots in Various Formats

Saving your visualizations is crucial when preparing figures for presentations, reports, or web publishing. Matplotlib provides a straightforward way to save your figures in a variety of formats.

#### Supported Formats:

- **Raster Formats:** PNG, JPEG (good for web use).
- **Vector Formats:** SVG, PDF (ideal for academic papers and professional reports because they can be scaled without losing quality).
- **Customization:** You can specify the resolution (DPI), size, and transparency in your saved plots.


````python
# Demonstrating the use of subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 12))  # A figure with 3 rows, each containing 1 plot
axs[0].plot(np.random.rand(50), 'r-')  # Red line plot
axs[1].bar(np.arange(10), np.random.rand(10))  # Bar chart
axs[2].scatter(np.arange(50), np.random.rand(50))  # Scatter plot
fig.suptitle('Subplots Example: Line, Bar, and Scatter', fontsize=16)
plt.show()
````

````python
# Saving a figure
fig, ax = plt.subplots()
ax.plot(np.cos(np.linspace(0, 2 * np.pi, 100)), 'g--')  # Green dashed line plot
fig.suptitle('Cosine Wave', fontsize=14)
fig.savefig('cosine_wave.pdf', format='pdf', dpi=300)  # Saving as a PDF with high resolution
plt.show()
````

# Part 2: Seaborn for Statistical Data Visualization

## Introduction to Seaborn

Seaborn is a powerful visualization library that builds on Matplotlib and provides a higher-level interface for drawing attractive and informative statistical graphics. It's particularly suited for visualizing complex datasets and is highly customizable, making it a valuable tool for both simple and sophisticated visualizations.

## Why Choose Seaborn Over Matplotlib?

Seaborn simplifies the process of creating many common visualization types:

- **Statistical routines:** Seaborn integrates closely with pandas data structures and provides built-in functions for statistical analysis, reducing the amount of code you need for complex visualizations.
- **Better aesthetics:** By default, Seaborn provides more aesthetically pleasing themes and color palettes to enhance the visual appeal and clarity of your plots.
- **Simplified syntax:** Seaborn has a more concise syntax compared to Matplotlib, making it easier to create complex visualizations.

## Visualization with Seaborn

Seaborn excels at creating a wide range of statistical plots. Let's explore some of the basic types of visualizations you can create with Seaborn.

### Distribution Plots

These plots help in visualizing the distribution of data, useful for understanding the underlying trends and patterns.
- **distplot**: Visualizes the distribution of a univariate set.
- **kdeplot**: Kernel density estimate plot.
- **histplot**: Histogram for viewing distributions.



````python
import seaborn as sns
import numpy as np
import pandas as pd

# Distribution plots
data = np.random.normal(size=100)
sns.histplot(data, kde=True)
plt.show()
````

### Categorical Plots

These plots are great for comparing categories with quantitative data.
- **boxplot**: Shows distributions with quartiles and outliers.
- **violinplot**: Combines boxplot and kernel density estimate.
- **swarmplot**: Categorical scatterplot with non-overlapping points.


````python
import pandas as pd

# Sample data
df = pd.DataFrame({
    "Category": ['A', 'B', 'A', 'B'],
    "Value": [10, 15, 10, 20]
})

# Creating a boxplot
sns.boxplot(x='Category', y='Value', data=df)
````

### Multi-plot Grids

Multi-plot grids are useful for plotting complex relationships across multiple variables.
- **pairplot**: Shows pairwise relationships across an entire dataframe.
- **FacetGrid**: Facilitates plotting different subsets of data.


````python
# Loading example dataset
iris = sns.load_dataset('iris')

# Creating a pairplot
sns.pairplot(iris, hue='species')
````

# Part 3: Integrating Matplotlib and Seaborn

In this section, we explore how to combine the strengths of both Matplotlib and Seaborn to enhance the quality and flexibility of our data visualizations. While Seaborn is built on top of Matplotlib and simplifies many plotting tasks, Matplotlib provides more control over plot elements. By integrating them, we can leverage the high-level interfaces of Seaborn with the detailed customization capabilities of Matplotlib.

## Combining Plots

Combining Matplotlib and Seaborn involves using Seaborn's high-level functions to create the plot and Matplotlib's extensive features to refine and customize the plot's details.

### Benefits of Integration

- **Flexibility**: Use Seaborn for easy setup and Matplotlib for fine-tuning.
- **Enhanced Functionality**: Combine statistical functions of Seaborn with the layout and annotation capabilities of Matplotlib.
- **Improved Aesthetics**: Start with Seaborn's appealing themes and use Matplotlib to adjust elements like labels, legends, and annotations for detailed customization.


### Examples of Combined Usage

Let's go through some examples where we create plots using Seaborn and then refine them using Matplotlib's functionalities.


````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Data preparation
data = np.random.normal(loc=0, scale=1, size=100)
df = pd.DataFrame(data, columns=['Data'])

# Create a basic histogram using Seaborn
plt.figure(figsize=(10, 6))
plot = sns.histplot(df['Data'], kde=True, color='blue')

# Customize using Matplotlib
plot.set_title('Histogram with Kernel Density Estimate', fontsize=20)
plot.set_xlabel('Values', fontsize=14)
plot.set_ylabel('Frequency', fontsize=14)
plt.grid(True)
plt.show()
````
